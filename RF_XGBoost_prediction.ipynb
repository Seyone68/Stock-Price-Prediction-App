## Stock Price Prediction using Random Forest and XGBoost with Stacking ensemble and Hyperparameter tuning

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
from pandas_datareader import data as pdr
import itertools
from sklearn.ensemble import RandomForestRegressor, StackingRegressor
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression, RidgeCV, LassoCV, ElasticNetCV
from xgboost import XGBRegressor

import warnings
warnings.filterwarnings('ignore')
```

```python
# Configure Yahoo Finance data source
# Download historical stock data
ticker = "GOOGL"
start_date = "2018-01-01"
end_date = "2023-10-31"

df = yf.download(ticker, start=start_date, end=end_date, progress=False)
df
```

```python
# Plot the closing price history
plt.figure(figsize=(16,8))
plt.title('Closing Price History', fontsize=18)
plt.plot(df['Close'], linewidth=2, color='black')
plt.xlabel('Date', fontsize=18)
plt.ylabel('Closing Price', fontsize=18)
plt.show()
```

```python
# Feature Engineering
df['Return'] = df['Close'].pct_change() # Simple daily return
df['Moving_Avg_5'] = df['Close'].rolling(window=5).mean() # 5-day moving average
df['Moving_Avg_10'] = df['Close'].rolling(window=10).mean() # 10-day moving average
df['Moving_STD_10'] = df['Close'].rolling(window=10).std() # 10-day moving standard deviation

# Replace NA values with 0
df.fillna(0, inplace=True)

df.head(30)
```

```python
# Removing first 9 rows since they contain 0 values
df1 = df[9:]

# Drop n/a values
df1.dropna(inplace=True)

df1.head()
```

```python
# Define the features and the target variable
features = ['Open', 'High', 'Low', 'Return', 'Moving_Avg_5', 'Moving_Avg_10', 'Moving_STD_10']

X = df1[features]
y = df1['Close']

# Feature Scaling
feature_scaler = MinMaxScaler()
X_scaled = feature_scaler.fit_transform(X)

target_scaler = MinMaxScaler()
y_scaled = target_scaler.fit_transform(y.values.reshape(-1,1))


# Splitting the dataset into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.2, shuffle=False, random_state=42)
```

```python
print("X_train shape: ", X_train.shape)
print("y_train shape: ", y_train.shape)
print("X_test shape: ", X_test.shape)
print("y_test shape: ", y_test.shape)
```

```python
# Random Forest Regressor Model without Hyperparameter Tuning
rf = RandomForestRegressor()
rf.fit(X_train, y_train)
pred_rf = rf.predict(X_test)

# XGBoost Regressor Model without Hyperparameter Tuning
xgb = XGBRegressor()
xgb.fit(X_train, y_train)
pred_xgb = xgb.predict(X_test)
```

```python
# Create a stacked model with Random Forest and XGBoost
stacked_pred = np.column_stack((pred_rf, pred_xgb))

# Train a final regressor based on the stacked predictions
final_regressor = LinearRegression().fit(stacked_pred, y_test)

# Predict on the test data
final_pred = final_regressor.predict(stacked_pred)
```

```python
# Denormalize the predicted values
pred_rf_inv = target_scaler.inverse_transform(pred_rf.reshape(-1,1))
pred_xgb_inv = target_scaler.inverse_transform(pred_xgb.reshape(-1,1))
pred_stack_inv = target_scaler.inverse_transform(final_pred.reshape(-1,1))
y_test_inv = target_scaler.inverse_transform(y_test.reshape(-1,1))
```

```python
# Calculate the evaluation metrics for Random Forest Regressor
rf_mse = mean_squared_error(y_test_inv, pred_rf_inv)
rf_rmse = np.sqrt(rf_mse)
rf_r2 = r2_score(y_test_inv, pred_rf_inv)
rf_mae = mean_absolute_error(y_test_inv, pred_rf_inv)

print('Random Forest Regressor Evaluation Metrics')
print('-------------------------------------------')
print('MSE: {:.4f}'.format(rf_mse))
print('RMSE: {:.4f}'.format(rf_rmse))
print('MAE: {:.4f}'.format(rf_mae))
print('R2 Score: {:.4f}'.format(rf_r2))
```
